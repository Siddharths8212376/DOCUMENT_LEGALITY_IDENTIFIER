```
all_legals = df_legals['LEGALS']
all_commons = df_commons['COMMONS']
all_com_legals = df_com_legals['COMMON-LEGALS']

```

```
# now we shall import the test document and tokenize it.
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
with open('test_doc.txt', mode='r') as test_file:
  document = test_file.read()
print(document)
# nltk.download('punkt')
# print(nltk.data.path)
```

```
import string

def get_filtered_set(document):
  stop_words = set(stopwords.words('english'))
  punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
  punc_less_doc = ""
  for word in document:
    for punc in punctuations:
      if punc in word:
        word = word.replace(punc, " ")
        
  words = word_tokenize(document)

  filtered_set = []
  # remove all punctuations
  # table = str.maketrans('', '', string.punctuation)
  # words = [w.translate(table) for w in words]

  # make everything lowercase
  words = [word.lower() for word in words]
  # remove everything that's not an alphabet
  words = [word for word in words if word.isalpha()]
  # remove all stop words
  for word in words:
    for punc in punctuations:
      if punc in word:
        word = word.replace(punc, " ")
    if word not in stop_words and len(word) > 1:
      filtered_set.append(word)
  # print(filtered_set)
  return filtered_set
```

```
filtered_set = get_filtered_set(document)
def get_prediction(filtered_set):

  total_score = 0.0
  division_factor = float(len(filtered_set))
  for each_word in filtered_set:
    if each_word in list(all_legals):
      total_score += list(df_legals.loc[df_legals['LEGALS'] == each_word]['SCORE'])[0]
    elif each_word in list(all_commons):
      total_score += list(df_commons.loc[df_commons['COMMONS'] == each_word]['SCORE'])[0]
    elif each_word in list(all_com_legals):
      total_score += list(df_com_legals.loc[df_com_legals['COMMON-LEGALS'] == each_word]['SCORE'])[0]
    else:
      total_score += 0.0
      division_factor -= 1
  print(total_score, " ", division_factor)
  prediction = total_score / division_factor
  # print(prediction)
  return prediction

prediction = get_prediction(filtered_set)
# print(filtered_set)
print(prediction)
```

```
with open('test_legal.txt') as legal_file:
  doc_format = legal_file.read()

set_of_words = get_filtered_set(doc_format)
print(set_of_words)
```

```
predict_legal = get_prediction(set_of_words)
print(predict_legal)
```

```
def is_legal(prediction):
  if prediction <= 0.3:
    return False
  else:
    return True
```

```
is_legal(predict_legal)
```
